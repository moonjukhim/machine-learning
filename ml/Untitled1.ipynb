```python
# 예제 코드
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model



# 데이터 적재
insurance = pd.read_csv("https://raw.githubusercontent.com/moonjukhim/machine-learning/main/ml/data/insurance.csv", delimiter=",")
insurance.head()

# 데이터 준비
#country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)
X = np.c_[insurance["age"]]
y = np.c_[insurance["charges"]]

# 데이터 시각화
insurance.plot(kind='scatter', x="age", y='charges')
plt.show()

# 선형 모델 선택
model = sklearn.linear_model.LinearRegression()

# 모델 훈련
model.fit(X, y)

# 키프로스에 대한 예측
X_new = [[40]]  # 40세의 보험 비용
print(model.predict(X_new)) # 출력 [[ 13474.78975274]]
```
